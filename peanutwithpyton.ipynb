# String characters and slicing: 
Characters are individual letters in a string,

and slicing is used to extract a part (substring) of the string using indexes.


```python
# I m r a n ‚Üí characters
# 0 1 2 3 4 ‚Üí indexing


name = "Imran"
# We get a character using its index number

print(name[0])    # first character of string at index 0
print(name[1])    # character at index number 1
print(name[-1])   # character at last index
print(name[-2])   # character before last index (n-1)

## Slicing

print(name[1:3])  # between index 1 and 3 (includes 1, excludes 3)
print(name[1:])   # prints everything after index 1 (includes 1)
print(name[:4])   # prints everything before index 4
print(name[0:5])   # normal ‚Üí Imran

#SKIP CHARACTER 
# string[start : end : step]

print(name[0:5:1])   # normal ‚Üí Imran (characters printed one after another)
print(name[0:5:2])   # skip 1 each time ‚Üí Ira (characters printed every 2nd position)
print(name[::2])     # whole string with step 2 ‚Üí Ira (prints every 2nd character)
print(name[1::2])    # start at 1, step 2 ‚Üí mrn     () (prints every 2nd character starting from index 1)
## It starts with ‚Äúm‚Äù, then jumps 2 index positions to reach ‚Äúa‚Äù, then jumps 2 more and goes out of range ‚Äî so it prints ‚Äúma‚Äù.

```

## Closure 

A closure is a special feature where:
an inner (nested) function remembers and uses variables from its outer function
even after the outer function has finished executing

```python
def counter():
    count = 0

    def inc():
        nonlocal count
        count += 1
        return count

    return inc

c = counter()
print(c())
print(c())
```

```python
def counter():
    count = 0

    def increment():
        nonlocal count
        count = count + 1
        return count

    return increment


increment = counter()  # assign variable

print(increment())
print(increment())
print(increment())
```

## Loops 
while loops and for loops 

```python
# WHILE LOOP ‚Äî BASIC

condition = True          # Boolean variable

while condition:          # loop runs while condition is True
    print("Condition is True")
    condition = False     # change condition ‚Üí loop stops

# WHILE LOOP ‚Äî COUNTER CONTROLLED

count = 0                 # starting value

while count < 5:          # loop runs until count becomes 5
    print("Count =", count)
    count += 1            # increment counter (important)

print("While loop finished")

# WHILE LOOP ‚Äî WITH BREAK

num = 0

while True:               # infinite loop
    print(num)
    if num == 3:          # stop condition inside loop
        break             # exits loop immediately
    num += 1

# WHILE LOOP ‚Äî WITH CONTINUE

x = 0

while x < 5:
    x += 1
    if x == 3:
        continue          # skip this iteration
    print("Value:", x)

# FOR LOOP ‚Äî LIST ITERATION

items = [1, 2, 3, 4, 5]   # list

for item in items:        # loop through each element
    print(item)

# FOR LOOP ‚Äî STRING ITERATION

name = "IMRAN"

for ch in name:           # loop through each character
    print(ch)

# FOR LOOP ‚Äî RANGE


for i in range(5):        # generates numbers 0‚Äì4
    print(i)


# range(start, stop)

for i in range(2, 6):     # 2 to 5
    print(i)


# range(start, stop, step)

for i in range(1, 10, 2): # step of 2
    print(i)

# FOR LOOP ‚Äî ENUMERATE (index + value)

data = ["A", "B", "C"]

for index, value in enumerate(data):   # gives index and value
    print(index, value)

# FOR LOOP ‚Äî NESTED

for i in range(3):        # outer loop
    for j in range(2):    # inner loop
        print(i, j)

# FOR LOOP ‚Äî BREAK


for i in range(10):
    if i == 4:
        break             # stop loop early
    print(i)

# FOR LOOP ‚Äî CONTINUE

for i in range(5):
    if i == 2:
        continue          # skip when i = 2
    print(i)

# FOR LOOP ‚Äî PASS (placeholder)

for i in range(3):
    pass                  # do nothing (used for future code)

# FOR LOOP ‚Äî WITH ELSE

for i in range(3):
    print(i)
else:
    print("Loop completed normally")   # runs if no break happens
```
